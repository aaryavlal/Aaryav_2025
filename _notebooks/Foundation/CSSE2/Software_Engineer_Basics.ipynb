 updateAnimationState(key) { // Checks what keys player is pressing and sets the animation and function accordingly
        switch (key) {
            case 'a':
            case 'd':
                this.state.animation = 'walk'; //Animation is set to walk if a or d is pressed
                GameEnv.playerAttack = false; // Player attack is set to false
                break;
            case 'w':
                if (this.state.movement.up == false) {
                this.state.movement.up = true; // Sets jump movment to true if W is pressed
                this.state.animation = 'jump'; //Sets animation to jump
                }
                GameEnv.playerAttack = false;
                break;
            case 's': 
                if ("a" in this.pressedKeys || "d" in this.pressedKeys) {  // Check if player is pressing a or d
                this.state.animation = 'run'; //sets animation to run
                }
                GameEnv.playerAttack = false; 
                break;
            case 'b':
                // Trigger attack once per key press, no holding allowed
            if (!this.state.isAttacking) {  // Check if player isn't already attacking
                this.state.animation = 'attack';  // Set the animation to attack
                GameEnv.playerAttack = true;     // Set player attack state to true
                this.state.isAttacking = true;   // Set the player as attacking

                // Reset the attack state after the animation ends (set it to 500ms for now)
                setTimeout(() => {
                    this.state.isAttacking = false;
                    GameEnv.playerAttack = false;
                }, 500); // 500ms delay for one attack per key tap (adjust timing based on animation)
            }
            break;
        default:
            this.state.animation = 'idle'; // Sets animation to idle if no keys are pressed
            GameEnv.playerAttack = false; //Attack is false if no keys are pressed
            break;
            }
    }
