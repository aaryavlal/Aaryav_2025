/**

//Setting animation for different Skibidi player actions, including attack
    updateAnimationState(key) {
        switch (key) {
            case 'a':
            case 'd':
                this.state.animation = 'walk';
                GameEnv.playerAttack = false;
                break;
            case 'w':
                if (this.state.movement.up == false) {
                this.state.movement.up = true;
                this.state.animation = 'jump';
                }
                GameEnv.playerAttack = false;
                break;
            case 's':
                if ("a" in this.pressedKeys || "d" in this.pressedKeys) {
                this.state.animation = 'run';
                }
                GameEnv.playerAttack = false;
                break;
            case 'b':
                this.state.animation = 'attack';  // Always trigger attack when b is pressed
                GameEnv.playerAttack = true;
                break;
            default:
                this.state.animation = 'idle';
                GameEnv.playerAttack = false;
                break;
        }
    }

```javascript

case 'b':
                // Trigger attack once per key press, no holding allowed
            if (!this.state.isAttacking) {  // Check if player isn't already attacking
                this.state.animation = 'attack';  // Set the animation to attack
                GameEnv.playerAttack = true;     // Set player attack state to true
                this.state.isAttacking = true;   // Set the player as attacking

                // Reset the attack state after the animation ends 
                setTimeout(() => {
                    this.state.isAttacking = false;
                    GameEnv.playerAttack = false;
                }, 500); // 500ms delay for one attack per key tap (adjust timing based on animation)
            }
            break;
        default:
            this.state.animation = 'idle';
            GameEnv.playerAttack = false;
            break;
            }
    }
```
The code ensures the player can only attack once per key press, preventing continuous attacks by holding the 'b' key.

#### Key Points:
- **Attack Check**: The attack triggers only if `this.state.isAttacking` is `false`, ensuring no repeated attacks while one is active.
- **Trigger Attack**: When 'b' is pressed, the animation is set to `'attack'`, and `GameEnv.playerAttack` is set to `true`. The player is marked as attacking (`this.state.isAttacking = true`).
- **Cooldown (`setTimeout`)**: A 500ms delay is applied using `setTimeout`, after which the player can attack again by resetting `this.state.isAttacking` and `GameEnv.playerAttack`.
- **Default Case**: Other keys set the player's state to `'idle'` and disable the attack.

This ensures one attack per key press with a 500ms cooldown.
