### Numbers

- `1.20`, `0.50`, `0.90`, `0.40`, `0.30` in the `updateJump()` method:

    ```javascript
    jumpHeightFactor = 1.20;
    jumpHeightFactor = 0.50;
    ```

- `100`, `250`, `900`, `10` in different parts of the file:

    ```javascript
    this.y = this.y - 100;
    this.y = this.y - 250;
    setTimeout(async () => { }, 900);
    this.x -= 10;
    ```

### Strings

- `"easy"`, `"super_easy"`, `"normal"`, `"hard"` in the `updateJump()` and `handlePlayerReaction()` methods:

    ```javascript
    if (GameEnv.difficulty === "easy") { }
    ```

- `"transform 0.5s"`, `"rotate(-90deg) translate(-26px, 0%)"`, and other strings used for transition effects:

    ```javascript
    this.canvas.style.transition = "transform 0.5s";
    ```

- `"walk"`, `"jump"`, `"run"`, `"attack"`, `"idle"` in the `updateAnimationState()` method:

    ```javascript
    this.state.animation = 'walk';
    ```

### Booleans

- `true`, `false` values used throughout the code, such as:

    ```javascript
    this.state.isDying == false;
    if (GameEnv.goombaBounce === true) { }
    ```

- Also in properties like `GameEnv.invincible` and `this.state.isAttacking`:

    ```javascript
    GameEnv.invincible = false;
    ```

### Arrays

- `GameEnv.levels` is an array accessed with `.indexOf()` and passed into `GameControl.transitionToLevel()`:

    ```javascript
    await GameControl.transitionToLevel(GameEnv.levels[GameEnv.levels.indexOf(GameEnv.currentLevel)]);
    ```

### JSON Objects

- `this.collisionData.touchPoints` is an example of a nested JSON object:

    ```javascript
    if (this.collisionData.touchPoints.this.top && this.collisionData.touchPoints.other.bottom) { }
    ```

- `this.state` is also an object containing properties like `movement` and `animation`:

    ```javascript
    this.state.movement = { up: false, down: false, left: false, right: false, falling: false };
    this.state.animation = 'idle';
    ```
